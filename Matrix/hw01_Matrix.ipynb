```python
class Matrix:
  
  def __init__(self, data):
    self.data = data
  
  def __repr__(self):
    return f"Matrix({self.data})"

  
  
  def __add__(self, other):
    
    '''
    print(len(self.data), len(other.data))
    print(len(self.data[0]),len(other.data[0]))
    '''

    #행렬의 원소의 개수가 다를 때
    cnt_a = 0
    cnt_b = 0
    for element in self.data:
      for x in element:
        cnt_a += 1
    print(cnt_a)
    for element in other.data:
      for x in element:
        cnt_b += 1
    
    #행렬의 사이즈가 다를 때(행, 열)
    assert len(self.data) == len(other.data) and len(self.data[0]) == len(other.data[0]), "크기가 다릅니다!"
    #행렬의 사이즈는 같지만 원소의 개수가 다를 때
    #ex A = [[1,2,3], [4,5,6]] B = [[1,2,3,4], [4,5,6]]
    assert cnt_a == cnt_b, "원소의 개수가 다릅니다!"

    #필요하면 삭제 필요
    #assert [i for i in range(len(self.data)) if len(self.data[i]) == len(other.data[i])]
    
    answer = []
    for i in range(len(self.data)):
      tmp = []
      for j in range(len(self.data[i])):
        tmp.append(self.data[i][j] + other.data[i][j])
      answer.append(tmp)
    return Matrix(answer)

  def __sub__(self, other):

    #행렬의 원소의 개수가 다를 때
    cnt_a = 0
    cnt_b = 0
    for element in self.data:
      for x in element:
        cnt_a += 1
    print(cnt_a)
    for element in other.data:
      for x in element:
        cnt_b += 1
    
    #행렬의 사이즈가 다를 때(행, 열)
    #ex A = [[1,2,3,], [4,5,6]] B = [[1,1], [2,2], [3,3]]
    assert len(self.data) == len(other.data) and len(self.data[0]) == len(other.data[0]), "크기가 다릅니다!"
    #행렬의 원소의 개수가 다를 때
    #ex A = [[1,2,3], [4,5,6]] B = [[1,2,3,4], [4,5,6]]
    assert cnt_a == cnt_b, "원소의 개수가 다릅니다!"

    '''
    #assert 수정 필요
    for i in range(len(self.data)):
      print(i)
      if len(self.data[i]) != len(other.data[i]):
        print(len(self.data[i]), len(other.data[i]))
    assert [i for i in range(len(self.data)) if len(self.data[i]) == len(other.data[i])]
    '''
    answer = []
    for i in range(len(self.data)):
      tmp = []
      for j in range(len(self.data[i])):
        tmp.append(self.data[i][j] - other.data[i][j])
      answer.append(tmp)
    return Matrix(answer)

  def __mul__(self, other):


    #곱셈이 불가능한 경우 (앞 행렬의 열의 개수 != 뒷 행렬의 행의 개수)
    assert len(self.data[0]) == len(other.data)

    #행렬 * 행렬
    if type(other) == Matrix:
      answer = [[0]* len(other.data[0]) for _ in range(len(self.data))]
      for i in range(len(self.data)):
        for j in range(len(other.data[0])):
          for k in range(len(self.data[0])):
            answer[i][j] += self.data[i][k] * other.data[k][j]
      return Matrix(answer)
    #스칼라 곱 (행렬 * scalar)
    elif type(other) is int:
      answer = []
      for i in range(len(self.data)):
        tmp = []
        for j in range(len(self.data[i])):
          tmp.append(self.data[i][j] * other)
        answer.append(tmp)
      return Matrix(answer)

  #scalar * 행렬    
  def __rmul__(self, other):
    answer = []

    for i in range(len(self.data)):
      tmp = []
      for j in range(len(self.data[i])):
        tmp.append(self.data[i][j] * other)
      answer.append(tmp)
    return Matrix(answer)






a = Matrix([[1]])

b = Matrix([[1,2,3,4]])

print(a * b)
```
